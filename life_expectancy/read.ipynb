### Agrupando e interpolando por paÃ­s

```python
grups = life_expectancy_df.groupby('Country')
grups
```

```python
def interpolar_grupo(grupo):
    grupo_interpolado = grupo.interpolate(method='linear', limit_direction='both')
    return grupo_interpolado

df = grups.apply(interpolar_grupo)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.reset_index(drop=True)
```

```python
colunas_numericas = df.select_dtypes(include='number').columns

medias = df.mean()
df[colunas_numericas] = df[colunas_numericas].fillna(df[colunas_numericas].mean())
```
